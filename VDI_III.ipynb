```python
import pandas as pd
import matplotlib.pyplot as plt
```

sumber data: https://www.kaggle.com/datasets/mariumfaheem666/top-rated-movies


```python
data = pd.read_csv('/content/Top_Rated_Movies.csv')
data.head()
```

Tujuan:


*   Memberikan informasi mengenai tren dan pola Top Rated Movies.
*   Membantu produser dalam membaca pasar.
*   Membantu penonton dalam mencari referensi fiml.

User:

*   Producer film.
*   Consumer film.

Fungsi:

Memberikan informasi terkait rating film kepada producer dan consumer.

Tone:

Santai dan Optimis.










```python
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')

data = data.sort_values(by='release_date')

plt.figure(figsize=(10,6))
plt.plot(data['release_date'], data['vote_average'], marker='o', linestyle='-', color='b')

plt.title('Tren Rating Film Berdasarkan Tanggal Rilis', fontsize=14)
plt.xlabel('Tanggal Rilis', fontsize=12)
plt.ylabel('Rata-rata Vote', fontsize=12)

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
