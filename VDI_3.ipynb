```python
import pandas as pd
import matplotlib.pyplot as plt
```

sumber data: https://www.kaggle.com/datasets/mariumfaheem666/top-rated-movies


```python
data = pd.read_csv('/content/Top_Rated_Movies.csv')
data.head()
```

Tujuan:


*   Memberikan informasi mengenai tren dan pola Top Rated Movies.
*   Membantu produser dalam membaca pasar.
*   Membantu penonton dalam mencari referensi fiml.

User:

*   Producer film.
*   Consumer film.

Fungsi:

Memberikan informasi terkait rating film kepada producer dan consumer.

Tone:

Santai dan Optimis.










```python
top_movies = data.nlargest(10, 'vote_average')

plt.figure(figsize=(12, 6))
plt.barh(top_movies['title'], top_movies['vote_average'], color='grey')

plt.title('10 Film Teratas Berdasarkan Rating')
plt.xlabel('Rata-rata Rating ')
plt.ylabel('Judul Film')

plt.grid(axis='x')
plt.show()
```
